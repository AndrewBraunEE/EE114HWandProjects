```python
#Instead of his features, we load our own features that we know of
#Saves a pickle dictionary of features, and an array of their values in respect to time.
import glob, os
#import ujson
import json
import pickle
from scipy import signal
%matplotlib inline
import scipy.io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#features_to_select = ["sF0", "sF1", "sF2", "sF3", "sF4", "SHR", "Energy"]
features_to_select = ['sF1', 'sF2', 'sF3', 'sF4', 'SHR', 'Energy']

labels = []
#Iterate files, generate list of mat files
files = glob.glob("../data/*.mat")
print(files)
voice_speakers = []
for file in files:
    #Create features
    #os.chdir("../data")
    data = scipy.io.loadmat(file, verify_compressed_data_integrity=False)
    features = []
    
    filename = os.path.basename(file)
    print(filename)
    local_features_to_select = features_to_select.copy()
    speak_samples = 0
    
    feature_chunks = 4
    
    for feature in local_features_to_select:
        feature_data = data[feature]
        feature_data = feature_data[~np.isnan(feature_data)]
        speak_samples = len(feature_data)
        #os.chdir("../processed_data")
        #Compute the weighted average then downsample (decimate)
        '''
        N = 2048
        #print(feature_data)
        if feature_data != []:
            x = np.convolve(feature_data, np.ones((N,))/N, mode='valid')
            cnt = 0
            x_decim = []
            for index in range(0, N, len(x)-1):
                x_decim.append(x[index])
                cnt += 1
        else:
            x = []
        '''
        feature_data = np.array_split(feature_data, feature_chunks)
        feat_processed = []
        for chunk in feature_data:
            if chunk.size > 0:
                #max_val = np.amax(chunk)
                max_val = np.average(chunk)
                if max_val != None:
                    feat_processed.append(max_val)
                else:
                    feat_processed.append(0)
            else:
                feat_processed.append(0) #Make Farsi complete NaN example work
        features.append(feat_processed)
    #Add extra information that the dictionary should have, but cant load directly from the .mat file
    
    local_features_to_select.append("filename")
    local_features_to_select.append("label")
    local_features_to_select.append("speak_samples")
    
    features.append(filename)
    if "english" in filename:
        labels.append(1)
        features.append(1)
    else:
        labels.append(0)
        features.append(0)
    features.append(np.repeat(speak_samples, feature_chunks))
    feature_dict = dict(zip(local_features_to_select, features))
    voice_speakers.append(feature_dict)
    pickle.dump(feature_dict, open("../processed_data/" + filename + ".pkl", 'wb'))
    #Create labels
        
pickle.dump(labels, open("../processed_data/" + "labels.pkl", 'wb'))
pickle.dump(voice_speakers, open("../processed_data/" + "voice_speakers.pkl", 'wb'))
```

```python
#Load data
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pickle
labels = pickle.load(open("../processed_data/labels.pkl", 'rb'))
_voice_speakers = pickle.load(open("../processed_data/voice_speakers.pkl", 'rb'))

test_size = 0.15

#voice_speakers and labels are guaranteed to be in line
#features_to_select = ["sF0", "sF1", "sF2", "sF3", "sF4", "SHR", "Energy", "speak_samples"]
features_to_select = ["sF1", "sF2", "sF3", "sF4", "SHR", "Energy"]
#features_to_select = ["SHR", "Energy", "speak_samples"]

all_features = []
labels = []
for feature_dict in _voice_speakers:
    feature_list = []
    labels.append(feature_dict['label'])
    for key in features_to_select:
        #Does not select label, or filename
        feature_list.append(feature_dict[key])
    #Create a matrix that is n features columns, sample times rows.
    np_feature_matrix = np.matrix(feature_list)
    all_features.append(np_feature_matrix)
    
#print(all_features)

X_train, X_test, y_train, y_test = train_test_split(all_features, labels, test_size = test_size, random_state = 26)
#X_train, X_test, y_train, y_test = train_test_split(MFCC_features, labels_mfcc, test_size = test_size, random_state = 26)
y_train, y_test = np.array(y_train), np.array(y_test)
X_train, X_test = np.array(X_train), np.array(X_test)
#print("X_test: " + str(X_test) + "y_test:" + str(y_test))
print("X_test: " + "[" + str(len(X_test)) + ", " + str((X_test[0].shape)) + "y_test: " + str(y_test))

#Each column contains the list of features, we will vectorize this after

print(y_train.shape)
print(len(X_train))
```

```python
#Run the neural network
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras import optimizers
from sklearn.preprocessing import OneHotEncoder
import sys
np.set_printoptions(threshold=sys.maxsize)

#Onehot encode labels for Keras model (mandatory)
onehot_encoder = OneHotEncoder(sparse=False, categories='auto')

_y_train = y_train.reshape(-1, 1)
_y_test = y_test.reshape(-1, 1)

_y_train = onehot_encoder.fit_transform(_y_train)
_y_test = onehot_encoder.fit_transform(_y_test)

#Vectorize X_train features
features_each_speaker = []
print(X_train)
print(_y_train)

#Changing the shape of this CNN doesn't seem to really affect the gradient descent of the data
#Converges at the same minima.
#Dropout and learning_rate makes no difference

modelNN = Sequential()

modelNN.add(Dense(39, activation='softmax', input_shape=(X_train[0].shape[0], X_train[0].shape[1])))
#model.add(Dense(24, activation='softmax', input_shape=(X_train[0].shape[0],X_train[0].shape[1])))
#model.add(Dense(64, activation='softmax'))
modelNN.add(Dropout(0.2))
modelNN.add(Dense(78, activation='softmax'))
modelNN.add(Dropout(0.2))
modelNN.add(Dense(10, activation='softmax'))
modelNN.add(Flatten())
modelNN.add(Dense(2, activation='softmax'))         

Adam=optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999)



modelNN.compile(loss='binary_crossentropy',
              optimizer=Adam,
              metrics=['accuracy'])

history=modelNN.fit(np.array(X_train), np.array(_y_train), validation_data=(X_test, _y_test), epochs=100)
#Neural Network achieved about 64% Accuracy using our model.
```
